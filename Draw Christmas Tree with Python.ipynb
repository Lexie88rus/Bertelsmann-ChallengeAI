# Draw a Christmas Tree with Python

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

I used this [cartesian art](https://www.google.com/imgres?imgurl=https://i.pinimg.com/originals/8c/84/5f/8c845f8dc0f633e29b3667b0c7a00113.jpg&imgrefurl=https://www.pinterest.com/pin/407575835006779859/&docid=LpxGHx4dOWNO_M&tbnid=i1sIZ-rFdm0elM:&vet=1&w=1224&h=1584&hl=en-us&source=sh/x/im) and this [color palette](https://www.color-hex.com/color-palette/1483) to create the Christmas tree and [this palette](https://www.color-hex.com/color-palette/84930) for the decorations:

```python
plt.figure(figsize=(10,10))

tree_edges = [[25,34,29,36,31,38,33,42,8,17,12,19,14,21,16,25],
              [50,40,40,30,30,20,20,7,7,20,20,30,30,40,40,50]]

# plot tree edges
plt.plot(tree_edges[0], tree_edges[1], color='#74d680')
# fill the tree with green
lower_edge = [7 for i in range(0, len(tree_edges[0]))]
plt.fill_between(x=tree_edges[0], y1=lower_edge, y2=tree_edges[1],
                 where= tree_edges[1] >= lower_edge, facecolor='#74d680')

# plot the stem of tree
stem_edges = [[30,30,20,20, 30],[0,7,7,0,0]]
plt.plot(stem_edges[0], stem_edges[1], color='#ba632e')
# fill the stem with brown
plt.fill_between(x=[20,30], y1=[0,0], y2=[7,7],
                 where= [7,7] >= [0,0], facecolor='#ba632e')

# plot the decorations
plt.plot([31,27,23,17], [44,43,42,41], color='#ff9e15', marker='o', markersize=12)
plt.plot([32,30,26,20,15], [36,34,33,32,32], color='#004c97', marker='s', markersize=12)
plt.plot([18,20,24,31,37], [29,27,25,23,22], color='#e01e5a', marker='*', markersize=12)
plt.plot([35,33,30,23,15,10], [17,15,13,11,10,10], color='#00c2c7', marker='H', markersize=12)

# add a star at the top
plt.scatter([25], [50], marker='*', s=1000, color = '#ffde47')

# add colored balls
plt.scatter([15,30,23],[14,28,38],s=[500, 700,600],color=['#ffde47', '#ffde47','#ffde47'])
plt.scatter([18,25,35],[23,30,12],s=[500, 700,600],color=['#e9021d', '#e9021d','#e9021d'],marker='p')
plt.scatter([23,29],[18,9],s=[500, 700],color=['#016b34', '#016b34'],marker='h')
plt.scatter([25,27,20],[45,22,11],s=[400, 500,500],color=['#ffffff', '#ffffff', '#ffffff'],marker='d')

# add a title
plt.title('Merry Christmas!', fontsize=20)
plt.xlabel('And a Happy New Year!', fontsize=20)

plt.show()
```
